```python
df = pd.read_csv('../data/raw/frame_1.csv')
df.head()
```

```python
df.info()
df.isna().sum()
```

```python
df['Data'] = pd.to_datetime(df['Data'], errors = 'coerce')
```

```python
tidy_df = pd.melt(
    df,
    id_vars = ['Placówka SG', 'Przejście', 'Rodzaj przejścia', 'Odcinek', 'Oddział SG', 'Data', 'Kto', 'Kierunek'],
    value_vars = ['Paszportowy', 'Pozasystemowa', 'MRG', 'Inny', 'Załogi pociągów osobowych', 'Załogi pociągów towarowych',
        'Załogi statków pasażerskich', 'Załogi statków handlowych',
        'Załogi statków rybackich', 'Załogi kutrów',
        'Załogi taboru rzecznego', 'Załogi jednostek sportowo-żeglarskich',
        'Załogi samolotów', 'Załogi śmigłowców', 'os. w INNYCH', 'Razem'
    ],
    var_name = 'Typ transportu',
    value_name = 'Liczba osób',
)
```

```python
tidy_df['Data_M'] = tidy_df['Data'].dt.to_period('M')
```

```python
tidy_df
```

```python
tidy_df.to_csv('../data/processed/completed_tabel_nr1.csv', index = False)
```

### Preparation for ML

```python
roud_df = tidy_df[tidy_df['Rodzaj przejścia'] == 'drogowe'].copy()
roud_df.head()
```

```python
roud_df = roud_df[roud_df['Liczba osób'] > 0]
```

```python
roud_df['log_liczba_osób'] = np.log(roud_df['Liczba osób'])
roud_df.head()
```

```python
roud_df['Data'] = pd.to_datetime(roud_df['Data'])
roud_df['day'] = roud_df['Data'].dt.day
roud_df['month'] = roud_df['Data'].dt.month
roud_df['weekday'] = roud_df['Data'].dt.weekday
roud_df['is_wekend'] = roud_df['weekday'] >= 5
```

```python
roud_df.to_csv('../data/processed/roud_border_tabel_1.csv', index = False)
```
